```python
import pandas as pd
```

```python
data = pd.read_table("./dump.all")
```

```python
timeInstance = dict()
for i in range(0,len(data),14):
    pos1 = data.iloc[i+8,0].split(" ")[2:4]
    pos2 = data.iloc[i+8+1,0].split(" ")[2:4]
    pos3 = data.iloc[i+8+2,0].split(" ")[2:4]
    pos4 = data.iloc[i+8+3,0].split(" ")[2:4]
    pos5 = data.iloc[i+8+4,0].split(" ")[2:4]
    timeInstance[data.iloc[i,0]] = [[pos1[0],pos2[0],pos3[0],pos4[0],pos5[0]],[pos1[1],pos2[1],pos3[1],pos4[1],pos5[1]]]
    
```

```python
import matplotlib.pyplot as plt
```

```python
Xtime = []
XCoor1 = []
XCoor2 = []
XCoor3 = []
XCoor4 = []
XCoor5 = []
YCoor1 = []
YCoor2 = []
YCoor3 = []
YCoor4 = []
YCoor5 = []
for key,val in timeInstance.items():
    Xtime.append(int(key)/10000)
    XCoor1.append(float(val[0][0]))
    XCoor2.append(float(val[0][1]))
    XCoor3.append(float(val[0][2]))
    XCoor4.append(float(val[0][3]))
    XCoor5.append(float(val[0][4]))
    YCoor1.append(float(val[1][0]))
    YCoor2.append(float(val[1][1]))
    YCoor3.append(float(val[1][2]))
    YCoor4.append(float(val[1][3]))
    YCoor5.append(float(val[1][4]))
    
```
