```python
time_Arr = []
x_Arr = []
y_Arr = []
for i in data.index:
    instance  = data.loc[i]
    time_Arr.append(i)
    center_x = (instance[42463][0]+instance[42491][0]+instance[42498][0])/3
    x_Arr.append(center_x)
    center_y = (instance[42463][1]+instance[42491][1]+instance[42498][1])/3
    y_Arr.append(center_y)


centroid_df =pd.DataFrame({"time":time_Arr,"x":x_Arr,"y":y_Arr})
```

```python
# y Vs x
plt.figure(figsize=(20,10),dpi=400)
plt.plot(x_Arr,y_Arr)
plt.ylabel('height (y)')
plt.xlabel('x-coordinate')
plt.title('Center of mass of intruder')
plt.legend()
plt.savefig('intruder.jpg',bbox_inches='tight')
plt.show()
```

```python
# x Vs y
plt.figure(figsize=(20,10),dpi=400)
plt.gca().invert_xaxis()
plt.plot(y_Arr,x_Arr)
plt.xlabel('height (y)')
plt.ylabel('x-coordinate')
plt.title('Center of mass of intruder')
plt.legend()
plt.savefig('intruder_xVsy.jpg',bbox_inches='tight')
plt.show()
```

```python
# y Vs time
plt.figure(figsize=(20,10),dpi=400)
plt.plot(time_Arr,y_Arr)
plt.xlabel('time')
plt.ylabel('y-coordinate')
plt.title('Center of mass of intruder')
plt.savefig('intruder y Vs time.jpg',bbox_inches='tight')
plt.show()
```
